```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
def func(data):
    for i in range(4):
        if i==0:
            df = data
            df1 = df[df["Ticker"].str.startswith("NIFTY")]
            df = df1[df1["Ticker"].str.contains("CE|PE")]
            data_pattern = r'(\d{2}[a-zA-Z]{3}\d{2})' 
            df1 = df["Ticker"].str.split(data_pattern,expand = True).rename(columns = {0:"scrip",1:"expiry date",2:"strike"})
            df["index"],df["expiry date"],df["scrip"] = df1["scrip"],df1["expiry date"],df1["strike"]
            df.drop(columns = "Ticker",inplace = True)
            df["expiry date"] = pd.to_datetime(df["expiry date"])
            df["Time"] = pd.to_timedelta(df["Time"])
            df["Date"] = pd.to_datetime(df["Date"])
            df["Date"] = df["Date"].dt.strftime("%d%m%y")
            df["Date"] = pd.to_datetime(df["Date"])
            df["date_time"] = df["Date"] + df["Time"]
            df["dtte"] = df["expiry date"] - df["Date"]
            df["dte"] = df["dtte"].dt.days.astype("int32")
            df.drop(columns = ["dtte","Date","Time","Volume","Open Interest"],inplace = True)
            df.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close"},inplace = True)
            df = df[["index","scrip","open","high","low","close","expiry date","dte","date_time"]]
            df['expiry date'] = pd.to_datetime(df['expiry date']).dt.date
            df.reset_index(drop = True,inplace = True)
            for i in df["dte"]:
                if i<0:
                    df.drop(index = np.where(df["dte"]==i),inplace = True)
                else:
                    pass
            df.sort_values(by = ["date_time","close"],inplace = True)
            df.to_csv("NIFTY OPTIONS.csv",index = False)
        elif i==1:
            df = data
            df1 = df[df["Ticker"].str.contains("BANKNIFTY")]
            df = df1[df1["Ticker"].str.contains("CE|PE")]
            data_pattern = r'(\d{2}[a-zA-Z]{3}\d{2})' 
            df1 = df["Ticker"].str.split(data_pattern,expand = True).rename(columns = {0:"scrip",1:"expiry date",2:"strike"})
            df["index"],df["expiry date"],df["scrip"] = df1["scrip"],df1["expiry date"],df1["strike"]
            df.drop(columns = "Ticker",inplace = True)
            df["expiry date"] = pd.to_datetime(df["expiry date"])
            df["Time"] = pd.to_timedelta(df["Time"])
            df["Date"] = pd.to_datetime(df["Date"])
            df["Date"] = df["Date"].dt.strftime("%d%m%y")
            df["Date"] = pd.to_datetime(df["Date"])
            df["date_time"] = df["Date"] + df["Time"]
            df["dtte"] = df["expiry date"] - df["Date"]
            df["dte"] = df["dtte"].dt.days.astype("int32")
            df.drop(columns = ["dtte","Date","Time","Volume","Open Interest"],inplace = True)
            df.rename(columns = {"Open":"open","High":"high","Low":"low","Close":"close"},inplace = True)
            df = df[["index","scrip","open","high","low","close","expiry date","dte","date_time"]]
            df['expiry date'] = pd.to_datetime(df['expiry date']).dt.date
            df.reset_index(drop = True,inplace = True)
            for i in df["dte"]:
                if i<0:
                    df.drop(index = np.where(df["dte"]==i),inplace = True)
                else:
                    pass
            df.sort_values(by = ["date_time","close"],inplace = True)
            df.to_csv("BANKNIFTY OPTION.csv",index = False)
        elif i==3:
            df = data
            df = df[df["Ticker"] == "BANKNIFTY-I"]
            df["Date"] = pd.to_datetime(df["Date"])
            df["Time"] = pd.to_timedelta(df["Time"])
            df["date_time"] = df["Date"] + df["Time"]
            df.sort_values(by = ["date_time","Close"],inplace = True)
            df.rename(columns = {"Ticker":"index","Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume","Open Interest":"open interest"},inplace = True)
            df.reset_index(drop = True)
            df = df[["index","open","high","low","close","date_time"]]
            df.to_csv("BANKNIFTY FUTURE.csv",index = False)
        else:
            df = data
            df = df[df["Ticker"]=="NIFTY-1"]
            df["Date"] = pd.to_datetime(df["Date"])
            df["Time"] = pd.to_timedelta(df["Time"])
            df["date_time"] = df["Date"] + df["Time"]
            df.sort_values(by = ["date_time","Close"],inplace = True)
            df.rename(columns = {"Ticker":"index","Open":"open","High":"high","Low":"low","Close":"close","Volume":"volume","Open Interest":"open interest"},inplace = True)
            df.reset_index(drop = True)
            df = df[["index","open","high","low","close","date_time"]]
            df = df.to_csv("NIFTY FUTURE.csv",index = False)
            
            
            
    
```
