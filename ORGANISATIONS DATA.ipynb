```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\bhawna\\Downloads\\GFDLNFO_27012020.csv")
```

# 1.1 MAKING NEW DFS NAMED BANKNIFTY BY FILTERING.

```python
Banknifty_data = df[df["Ticker"].str.contains("BANKNIFTY")]
```

```python
Banknifty_data.head()
```

# 1.2 FILTERING DATAFRAME BANKNIFTY THAT END WITH "CE" AND "PE"

```python
Banknifty_data = Banknifty_data[Banknifty_data["Ticker"].str.contains("CE|PE")]
```

```python
Banknifty_data.head()
```

# 1.3 DROPPING VOLUME AND OPEN INTEREST COLUMNS

```python
Banknifty_data.drop(columns = ["Volume","Open Interest"],inplace = True)
```

# 1.4 EXTRACTING EXPIRY DATE AND STRIKE

```python
import re
```

```python
data_pattern = r'(\d{2}[a-zA-Z]{3}\d{2})'
```

```python
df1 = Banknifty_data["Ticker"].str.split(data_pattern,expand  = True).rename(columns = {0:"scrip",1:"expiry_date",2:"strike"})
```

```python
df1.head()
```

```python
Banknifty_data["Scrip"],Banknifty_data["Expiry_date"],Banknifty_data["Strike"] = df1["scrip"],df1["expiry_date"],df1["strike"]
```

```python
Banknifty_data.head()
```

# 1.5 DROPPED SCRIP COLUMN 

```python
Banknifty_data.drop(columns = "Ticker",inplace = True)
```

```python
Banknifty_data = Banknifty_data[["Scrip","Strike","Date","Expiry_date","Time","Open","High","Low","Close"]]
```

```python
Banknifty_data.head()
```

```python
import datetime
```

```python
Banknifty_data["Expiry_date"] = pd.to_datetime(Banknifty_data["Expiry_date"])
```

```python
Banknifty_data["Date"] = pd.to_datetime(Banknifty_data["Date"])
```

```python
Banknifty_data.head()
```

```python
Banknifty_data.dtypes
```

```python
Banknifty_data["dtte"] = Banknifty_data["Expiry_date"] - Banknifty_data["Date"]
```

```python
Banknifty_data.head()
```

```python
import datetime as dt
```

# 1.6 CREATED A COLUMN CALLED 'dte'

```python
Banknifty_data["dte"] = Banknifty_data["dtte"].dt.days.astype("int32")
```

```python
Banknifty_data.head()
```

```python
Banknifty_data.drop(columns = "dtte",inplace = True)
```

# 1.7 Dropped all the rows having negative values of dte

```python
for i in Banknifty_data["dte"]:
    if i<0:
        col_index = np.where(Banknifty_data["dte"]==i)
        Banknifty_data.drop(index = col_index,inplace = True)
    else:
        pass
```

```python
Banknifty_data.shape
```

```python
Banknifty_data["dte"].min()
```

# 1.9 Filtered put rows which contain minimum dte value

```python
Banknifty_data.iloc[np.where(Banknifty_data["dte"]==3)]
```

```python
Banknifty_data["Time"] = pd.to_timedelta(Banknifty_data["Time"])
```

```python
Banknifty_data.dtypes
```

```python
Banknifty_data["date_time"] = Banknifty_data['Date'] + Banknifty_data['Time']
```

```python
Banknifty_data.sort_values("date_time",inplace = True)
```
